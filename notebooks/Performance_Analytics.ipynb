# Top 5 Mutual Funds

```python
score = performance[[
    "scheme_name",
    "return_3yr_pct",
    "sharpe_ratio",
    "alpha",
    "expense_ratio_pct",
    "max_drawdown_pct"
]].copy()
```

```python
score["ReturnRank"] = score["return_3yr_pct"].rank(ascending=False)

score["SharpeRank"] = score["sharpe_ratio"].rank(ascending=False)

score["AlphaRank"] = score["alpha"].rank(ascending=False)

score["ExpenseRank"] = score["expense_ratio_pct"].rank(ascending=True)

score["DrawdownRank"] = score["max_drawdown_pct"].rank(ascending=False)
```

```python
score["Score"] = (
    score["ReturnRank"] * 0.30 +
    score["SharpeRank"] * 0.25 +
    score["AlphaRank"] * 0.20 +
    score["ExpenseRank"] * 0.15 +
    score["DrawdownRank"] * 0.10
)
```

```python
score = score.sort_values("Score")
```

```python
top5 = score.head(5)

top5[["scheme_name", "Score"]]
```

```python
top5 = score.head(5)

top5_codes = performance[
    performance["scheme_name"].isin(top5["scheme_name"])
][["amfi_code", "scheme_name"]]

top5_codes
```

```python
top5_nav = nav[
    nav["amfi_code"].isin(top5_codes["amfi_code"])
]

top5_nav.head()
```

```python
top5_nav = top5_nav.merge(
    top5_codes,
    on="amfi_code"
)

top5_nav.head()
```

```python
top5_nav["rebased_nav"] = top5_nav.groupby("scheme_name")["nav"].transform(
    lambda x: x / x.iloc[0] * 100
)
```

```python
import plotly.express as px

fig = px.line(
    top5_nav,
    x="date",
    y="rebased_nav",
    color="scheme_name",
    title="Top 5 Mutual Funds Performance"
)

fig.show()
```

```python
benchmark["rebased_value"] = benchmark.groupby("index_name")["close_value"].transform(
    lambda x: x / x.iloc[0] * 100
)
```

```python
benchmark_plot = benchmark.rename(
    columns={
        "index_name": "scheme_name",
        "rebased_value": "rebased_nav"
    }
)
```

```python
benchmark_plot = benchmark_plot[
    ["date", "scheme_name", "rebased_nav"]
]
```

```python
fund_plot = top5_nav[
    ["date", "scheme_name", "rebased_nav"]
]
```

```python
comparison = pd.concat(
    [fund_plot, benchmark_plot],
    ignore_index=True
)
```

```python
import plotly.express as px

fig = px.line(
    comparison,
    x="date",
    y="rebased_nav",
    color="scheme_name",
    title="Top 5 Mutual Funds vs NIFTY50 & NIFTY100"
)

fig.show()
```

```python
fig.write_html("benchmark_comparison.html")
```
